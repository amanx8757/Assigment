Cannot be Changed: Once created, the objectâ€™s contents cannot be modified.

4. Discuss the different types of operators in Python and provide examples of how they are used.

 A. Arithmetic Operators
These perform basic mathematical operations.

+ (Addition): Adds two numbers.
- (Subtraction): Subtracts one number from another.
* (Multiplication): Multiplies two numbers.
/ (Division): Divides one number by another.
% (Modulus): Returns the remainder of division.
** (Exponentiation): Raises one number to the power of another.
// (Floor Division): Divides and returns the largest integer smaller than or equal to the result.

Example:
   x = 10
   y = 3
   print(x + y)   # 13
   print(x - y)   # 7
   print(x * y)   # 30
   print(x / y)   # 3.33
   print(x % y)   # 1
   print(x ** y)  # 1000 (10 raised to the power of 3)
   print(x // y)  # 3 (floor division)


   B. Comparison Operators
  These compare two values and return True or False.

  == (Equal): True if values are equal.
  != (Not Equal): True if values are not equal.
  > (Greater Than): True if left is greater than right.
  < (Less Than): True if left is less than right.
  >= (Greater Than or Equal): True if left is greater than or equal to right.
  <= (Less Than or Equal): True if left is less than or equal to right.

Example:
  x = 10
  y = 3
  print(x == y)  # False
  print(x != y)  # True
  print(x > y)   # True

  C.Logical Operators
These are used to combine conditional statements.

and: True if both statements are true.
or: True if at least one statement is true.
not: Reverses the result (True becomes False, False becomes True).

Example:
  x = 10
  y = 3
  print(x > 5 and y < 5)  # True (both conditions are true)
  print(x > 15 or y < 5)  # True (one condition is true)
  print(not(x == y))      # True (x is not equal to y)

D. Assignment Operators
   These assign values to variables.

   = (Assign): Assigns a value.
   += (Add and Assign): Adds and assigns.
   -= (Subtract and Assign): Subtracts and assigns.
   *= (Multiply and Assign): Multiplies and assigns.
   /= (Divide and Assign): Divides and assigns.

Example:
   x = 5
   x += 3  # x = x + 3
   print(x)  # 8
   x *= 2  # x = x * 2
   print(x)  # 16

E.  Bitwise Operators
These work on bits and perform bit-by-bit operations.

& (AND): Sets each bit to 1 if both bits are 1.
| (OR): Sets each bit to 1 if one of two bits is 1.
^ (XOR): Sets each bit to 1 if only one of the bits is 1.
~ (NOT): Inverts all the bits.
<< (Left Shift): Shifts bits left, adding zeroes on the right.
>> (Right Shift): Shifts bits right, discarding bits on the right.

Example:
  x = 5   # 0101 in binary
  y = 3   # 0011 in binary
  print(x & y)  # 1 (0001 in binary)
  print(x | y)  # 7 (0111 in binary)
  print(x ^ y)  # 6 (0110 in binary)
  print(~x)     # -6 (Inverts bits of 5)


E.Identity Operators
These compare the memory location of two objects.

is: True if both variables point to the same object.
is not: True if they do not point to the same object.

Example:
   x = [1, 2, 3]
   y = [1, 2, 3]
   print(x is y)      # False (because x and y are different objects)
   print(x is not y)  # True


F.Membership Operators
These test if a value is part of a sequence (like a list, string, etc.).

in: True if value is found in the sequence.
not in: True if value is not found in the sequence.

Example:
   x = [1, 2, 3, 4]
   print(3 in x)     # True
   print(5 not in x) # True


5. Explain the concept of type casting in Python with examples

--> Type casting in Python means changing the data type of a value from one type to another. This can be done automatically by Python is implicit or manually by you is explicit.

Implicit Type Casting:

Python automatically converts data types when needed.

example:

x = 5       # int
y = 2.5     # float
z = x + y   # Python converts `x` to float
print(z)    # Output: 7.5

Explicit Type Casting:

You can manually change the type using built-in functions like int(), float(), or str().

example:
x = "10"
y = int(x)  # Converts string to integer
print(y)    # Output: 10

-- In Python, loops are used to execute a block of code repeatedly as long as a condition is true. There are two main types of loops:

  A. for loop
The for loop is used to iterate over a sequence (like a list, tuple, string, or range). It goes through each item in the sequence one by one.

  Use Cases:
  When you know how many times you want to repeat the block of code.
   Iterating through elements of a list, string, or other collections.

  Example:

   fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

 B. while loop
The while loop keeps executing the block of code as long as the condition remains true.

Use Cases:
When you donâ€™t know in advance how many times the loop should run.
When the loop needs to run until a condition is met.

 Example:
  i = 1
  while i <= 5:
    print(i)
    i += 1

c. break and continue Statements
break: Used to exit the loop early when a condition is met.
continue: Skips the current iteration and moves to the next one

Example:

 for i in range(1, 6):
    if i == 3:
        continue  # Skips when i is 3
    if i == 5:
        break  # Stops the loop when i is 5
    print(i)
